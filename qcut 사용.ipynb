# 동일한 구간으로 값을 나누기 

```python
import numpy as np
import pandas as pd
```

```python
# 연습용 데이터 

lst = list(range(1,101))
df= pd.DataFrame(lst, columns = ['d1'])  # 동일한 1~100 까지 1개 씩 값을 가지고 있음 
df['d2']= df['d1'].apply(lambda x: 1 if x < 50 else x) #  1 이 50개, 51~100까지 1개의 값을 가지고 있음 
df['d3']= df['d1'].apply(lambda x: 1 if x < 75 else x) #  1 이 25개, 26~100까지 1개의 값을 가지고 있음 
df
```

##구간화를 하는 2가지 함수 
- qcut() : 분위수를 기반으로 나눠줌 
- cut() : 등구간으로 나눠줌

둘다 비슷한 사용법. 그렇지만 quct을 좀더 선호하는 이유는 중복값이 많아 구간이 정의한 값보다 적게 구간이 생성될 경우 qcut은 중간에 값이 빠진 구간이 없이 만들어줌.

```python
# 중복값이 없는경우 (Nice 함)
pd.qcut(df['d1'], 4)
```

```python
# 중복값이 많아 구간을 원하는 만큼 못 만들경우 (에러 발생)
pd.qcut(df['d2'], 4).value_counts()
```

```python
# 중복값이 많아 구간을 원하는 만큼 못 만들경우 
pd.qcut(df['d2'], 4, duplicates ='drop').value_counts(sort ='False')
```

```python
pd.qcut(df['d3'], 4, duplicates ='drop').value_counts(sort ='False')
```

```python
# cut 함수를 이용할 경우 
pd.cut(df['d2'], 4, duplicates ='drop').value_counts(sort ='False')
```

```python
pd.cut(df['d3'], 4, duplicates ='drop').value_counts(sort ='False')
```

```python
pd.cut(df['d3'], 4, duplicates ='drop').value_counts(sort ='False')
```
